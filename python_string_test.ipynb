아래 코드의 실행 결과를 예상해보세요.

>> string = 'abcd'

>> string.replace('b', 'B')

>> print(string)

```python
# 기존 문자열이 그대로 출력이 된다.
# 문자열은 변경이 안되므로 변경된 문자열을 보려면
# 새로운 객체 문자열을 생성한 print(string.replace("b","B"로 봐야한다.))

string = 'abcd'

string.replace('b', 'B')

print(string)
```

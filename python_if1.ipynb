아래 코드의 출력 결과를 예상하라

    if True :
        if False:
            print("1")
            print("2")
        else:
            print("3")
    else :
        print("4")
    print("5")

```python
if True :
    if False:
        print("1")
        print("2")
    else:
        print("3")
else :
    print("4")
print("5")
# True -> True이므로 else 3 출력 -> 분기문 빠져나와서 5
```

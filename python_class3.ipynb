```python
# personal sol
import random

class Account():

  acc_count = 0
  deposit_acc = 0
  def __init__(self, name, balance):
    self.name = name
    self.balance = balance
    self.bank = "SC bank"
    
    num1 = random.randint(0, 999)
    num2 = random.randint(0, 99)
    num3 = random.randint(0, 999999)
    num1 = str(num1).zfill(3)      
    num2 = str(num2).zfill(2)      
    num3 = str(num3).zfill(6)
    self.acc_num = num1+"-"+num2+"-"+num3
    Account.acc_count += 1
    self.withdraw_log = []
    self.deposit_log = []

  

  @classmethod
  
  def get_account_num(cls):
    print(cls.acc_count)

# personal solution
  def deposit(self):
    
    print("현재 잔고 {}원".format(self.balance))
    while True:
      deposit = int(input())
      if deposit <= 0:
         self.balance += (deposit * 0)
         print("1원 이상 입력하세요")
         print(deposit)
      else: 
         self.balance += deposit
         Account.deposit_acc += 1
         self.deposit_log.append(deposit)
         print("입금 후 잔고 {}원".format(self.balance))
         print("입금횟수 {}".format(Account.deposit_acc))
         
         if Account.deposit_acc % 5 == 0:
           self.balance = (self.balance * 1.01)
           print("""입금을 5번 하셨으므로 1% 의 이자가 추가됩니다.
현재 잔고 {}원 """.format(self.balance))
      break

  def deposit_1(self, amount):
       if amount >= 1:
           self.balance += amount


  def withdraw(self):
    print("현재 잔고 {}원".format(self.balance))
    while True:
      withdraw = int(input())
      if withdraw > self.balance:
        print("not enough mineral")
      else:
        self.balance -= withdraw
        self.withdraw_log.append(withdraw)
        print("출금 후 잔고 {}원".format(self.balance))
        break
  def withdraw_1(self,amount):
    if amount < self.balance:
      self.balance -= amount
      print(self.balance)

  def display_info(self):
    print("은행이름 : {}".format(self.bank))
    print("예금주 : {}".format(self.name))
    print("계좌번호 : {}".format(self.acc_num))
    print("잔고 : {}".format(self.balance))

  def withdraw_history(self):
      for i in self.withdraw_log:
        print(i)

  def deposit_history(self):
      for i in self.deposit_log:
        print(i)


client_list = []
Oh = Account("yunseok",10000000)
Lee = Account("esul", 10000)
Jang = Account("yeonju", 10000)

client_list.append(Oh)
client_list.append(Lee)
client_list.append(Jang)

for i in client_list:
  if i.balance>=1000000:
    print(i.display_info())




# sol)
# import random


# class Account:
#     # class variable
#     account_count = 0

#     def __init__(self, name, balance):
#         self.deposit_count = 0
#         self.deposit_log = []
#         self.withdraw_log = []

#         self.name = name
#         self.balance = balance
#         self.bank = "SC은행"

#         # 3-2-6
#         num1 = random.randint(0, 999)
#         num2 = random.randint(0, 99)
#         num3 = random.randint(0, 999999)

#         num1 = str(num1).zfill(3)  # 1 -> '1' -> '001'
#         num2 = str(num2).zfill(2)  # 1 -> '1' -> '01'
#         num3 = str(num3).zfill(6)  # 1 -> '1' -> '0000001'
#         self.account_number = num1 + '-' + num2 + '-' + num3  # 001-01-000001
#         Account.account_count += 1


#     def get_account_num(cls):
#         print(cls.account_count)  # Account.account_count

#     def deposit(self, amount):
#         if amount >= 1:
#             self.deposit_log.append(amount)
#             self.balance += amount

#             self.deposit_count += 1
#             if self.deposit_count % 5 == 0:         # 5, 10, 15
#                 # 이자 지금
#                 self.balance = (self.balance * 1.01)


#     def withdraw(self, amount):
#         if self.balance > amount:
#             self.withdraw_log.append(amount)
#             self.balance -= amount

#     def display_info(self):
#         print("은행이름: ", self.bank)
#         print("예금주: ", self.name)
#         print("계좌번호: ", self.account_number)
#         print("잔고: ", self.balance)

    # def withdraw_history(self):
    #     for amount in self.withdraw_log:
    #         print(amount)

    # def deposit_history(self):
    #     for amount in self.deposit_log:
    #         print(amount)


k = Account("Kim", 1000)
k.deposit()
k.deposit()
k.deposit()
k.deposit_history()
print()
k.withdraw()
k.withdraw()
k.withdraw_history()
```
